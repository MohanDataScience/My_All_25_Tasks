```python
data=data.replace(['\$',''],'',regex=True)
data
```

```python
data['Total Net Worth']=data['Total Net Worth'].replace('B','',regex=True)
data
```

```python
data['Total Net Worth']=pd.to_numeric(data['Total Net Worth'],errors='coerce')
data
```

```python
data['$ YTD Change'].values
```

```python
def Values_to_float(x):
    if 'K' in x:
        return float(x.replace('K',''))*10
    if 'M' in x:
        return float(x.replace("M",''))*100
    if 'B' in x:
        return float(x.replace('B',''))*1000
data['$ Last Change']=data['$ Last Change'].apply(Values_to_float)
```

```python
data
```

```python
data['$ YTD Change']=data['$ YTD Change'].apply(Values_to_float)
data
```
